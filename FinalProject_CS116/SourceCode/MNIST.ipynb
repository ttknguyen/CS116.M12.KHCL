```python
from sklearn.datasets import fetch_openml
from sklearn.utils import shuffle
import numpy as np
```

```python
mnist = fetch_openml('mnist_784')
```

```python
X, y = np.float32(mnist.data[:70000]), np.float32(mnist.target[:70000])
X, y = shuffle(X,y)
#keeping just 15k training samples due to kPCA memory requirements
```

```python
train_img, train_lbl=np.float32(X[:15000]), np.float32(y[:15000])
```

```python
test_img, test_lbl=np.float32(X[60000:]), np.float32(y[60000:])
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
# Fit on training set only.
scaler.fit(train_img)
# Apply transform to both the training set and the test set.
train_img = scaler.transform(train_img)
test_img = scaler.transform(test_img)
```

```python
from sklearn.decomposition import KernelPCA
```

```python
# rbf_pca = KernelPCA(n_components = 300, kernel="rbf", gamma=0.04)
# X_reduced = rbf_pca.fit_transform(X)
```

```python
from sklearn.model_selection import GridSearchCV
from sklearn.linear_model import LogisticRegression
from sklearn.pipeline import Pipeline
```

```python
clf = Pipeline([
 ("kpca", KernelPCA()),
 ("log_reg", LogisticRegression())
 ])

param_grid = [
 {"kpca__kernel": ["linear"], "kpca__n_components": [300, 500, 700]},

 {"kpca__kernel": ["rbf"], "kpca__gamma": np.linspace(0.03, 0.05, 10), "kpca__n_components": [300, 500, 700]},

 {"kpca__kernel": ["sigmoid"], "kpca__gamma": np.linspace(0.03, 0.05, 10), "kpca__n_components": [300, 500, 700]},

 { "kpca__kernel": ["poly"], "kpca__degree": [1, 2, 3, 4, 5, 10], 
  "kpca__gamma": np.linspace(0.03, 0.05, 10), "kpca__n_components": [300, 500, 700]}
 ]
```

```python
grid_search = GridSearchCV(clf, param_grid, cv=3, verbose=3)
grid_search.fit(train_img, train_lbl)
```

```python
 print(grid_search.best_params_)
```

```python
import pandas as pd
pd.concat([pd.DataFrame(grid_search.cv_results_["params"]),pd.DataFrame(grid_search.cv_results_["mean_test_score"], columns=["Accuracy"])],axis=1)
```

```python
df = pd.concat([pd.DataFrame(grid_search.cv_results_["params"]),pd.DataFrame(grid_search.cv_results_["mean_test_score"], columns=["Accuracy"])],axis=1)
df = pd.concat([df, pd.DataFrame(grid_search.cv_results_["mean_fit_time"], columns=["Fit_time"])], axis=1)
```

```python
df.to_csv('file_mnist_gridsearch.csv', encoding='utf-8')
```

```python
grid_search.best_params_
```

```python
from sklearn.decomposition import PCA, KernelPCA
# Make an instance of the Model
pca = KernelPCA(n_components=700, gamma=grid_search.best_params_['kpca__gamma'], kernel=grid_search.best_params_['kpca__kernel'], degree=2)
```

```python
pca.fit(train_img)
```

```python
train_img = pca.transform(train_img)
test_img = pca.transform(test_img)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
# all parameters not specified are set to their defaults
# default solver is incredibly slow which is why it was changed to 'lbfgs'
logisticRegr = LogisticRegression(solver = 'lbfgs')
```

```python
logisticRegr.fit(train_img, train_lbl)
```

```python
# Predict for One Observation (image)
logisticRegr.predict(test_img[0].reshape(1,-1))
```

```python
# Predict for One Observation (image)
logisticRegr.predict(test_img[0:10])
```

```python
logisticRegr.score(test_img, test_lbl)
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
# Fit on training set only.
scaler.fit(train_img)
# Apply transform to both the training set and the test set.
train_img = scaler.transform(train_img)
test_img = scaler.transform(test_img)
```

```python
from sklearn.decomposition import KernelPCA
```

```python
from sklearn.model_selection import GridSearchCV
from sklearn.linear_model import LogisticRegression
from sklearn.pipeline import Pipeline
```

```python
from sklearn.metrics import mean_squared_error
def my_scorer(estimator, X, y=None):
  X_reduced = estimator.transform(X)
  X_preimage = estimator.inverse_transform(X_reduced)
  return -1 * mean_squared_error(X, X_preimage)
```

```python
clf = Pipeline([
 ("kpca", KernelPCA(n_components=500)),
 ("log_reg", LogisticRegression())
 ])
```

```python
param_grid = [{
 "gamma": np.linspace(0.03, 0.05, 10),
 "kernel": ["rbf", "sigmoid"]
 }]
```

```python
kpca=KernelPCA(fit_inverse_transform=True) 
grid_search = GridSearchCV(kpca, param_grid, cv=3, scoring=my_scorer, verbose=3)
grid_search.fit(X=train_img, y=None, groups=None)
```

```python
grid_search = GridSearchCV(kpca, param_grid, cv=3, verbose=3)
grid_search.fit(train_img)
```
