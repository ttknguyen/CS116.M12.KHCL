# Excersise 1: Hyperparameter of SVM

Hyperparameter ảnh hưởng nhiều đến việc xây dựng:
- **Kernel**: Hàm số dùng để thực hiện dự đoán\
Gồm các dạng hàm số: linear, poly, rbf, sigmoid, precomputed. Mặc định là rbf

- **Regularization (c)**: float, default=1.0.\
Điều chỉnh việc có nên bỏ qua các điểm dữ liệu bất thường trong quá trình tối ưu mô hình SVM.
- **Gamma**: ‘scale’, ‘auto’ or float, default=’scale’.\
Tham số gamma xác định việc sử dụng bao nhiêu điểm dữ liệu cho việc xây dựng siêu phẳng phân cách\


Ngoài ra còn có những hyperparameter khác do sklearn cung cấp để tối ưu quá trình huấn luyện model svm:
- coef: float, default=0.0
- shrinking: bool, default=True
- probability: bool, default=False
- tol: float, default=1e-3
- cache_size: float, default=200
- class_weight: dict or ‘balanced’, default=None
- verbose: bool, default=False
- max_iter: int, default=-1
- decision_function_shape: {‘ovo’, ‘ovr’}, default=’ovr’
- break_ties: bool, default=False
- random_state: int, RandomState instance or None, default=None
